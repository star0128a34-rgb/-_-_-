# Preprocess Past Regis of Major and Elect

```python
total_regis = pd.read_csv('./Data/Preprocess/total_regis_1.csv')
```

```python
total_regis_past = pd.read_csv('./Data/Student/total_regis_past.csv')
```

```python
total_regis_past['수강'] = 1
```

```python
total_regis = total_regis.merge(total_regis_past, how = 'left', on = '학수번호')
```

```python
total_regis['수강'] = total_regis['수강'].fillna(0)
```

```python
total_regis['수강'] = total_regis['수강'].astype('int')
```

# Preprocess Past Track of Elect

```python
ged_track_past = pd.read_csv('./Data/Student/ged_track_past.csv')
```

```python
ged_require = (ged_track_past['이수'].sum() < 3)
```

```python
ged_require_list = ged_track_past.loc[ged_track_past['이수'] == 0]['트랙'].tolist()
```

```python
def get_require(x):
    value = 0

    if ged_require:

        if x in ged_require_list:
            value = 1

    return value
```

```python
ged_regis = total_regis.loc[total_regis['이수구분'] == 15]
not_ged_regis = total_regis.loc[total_regis['이수구분'] != 15]
```

```python
ged_regis['필수'] = ged_regis['트랙'].apply(get_require)
```

```python
total_regis = pd.concat([not_ged_regis, ged_regis])
```

```python
total_regis.to_csv('./Data/Preprocess/total_regis_2.csv', index = 0)
```
